```python
import tensorflow as tf
from tensorflow.keras import layers, neural_models
from tensorflow.keras.datasets import cifar10
```

```python
# Load dataset
(X_train_data, Y_train_data), (X_test_data, Y_test_data) = cifar10.load_data()

# Normalize images (0-255 -> 0-1)
X_train_data = X_train_data / 255.0
X_test_data = X_test_data / 255.0
```

```python
# Data Augmentation
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1)
])
```

```python
# Build CNN Model
neural_model = neural_models.Sequential([
    data_augmentation,

    layers.Conv2D(16, (3,3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D(2, 2),

    layers.Dropout(0.3),

    layers.Flatten(),

    layers.Dense(10, activation='softmax')
])
```

```python
# Compile
neural_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train
neural_model.fit(
    X_train_data,
    Y_train_data,
    epochs=10,
    validation_data=(X_test_data, Y_test_data)
)
```
